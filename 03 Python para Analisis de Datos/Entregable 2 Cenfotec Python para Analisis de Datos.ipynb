```python
#Relacion de Tiros a puerta y fuera de puerta.

plt.figure(figsize=(10, 6))
plt.scatter(df['Shots on target'], df['Shots off target'],color='blue')
plt.title('Tiros a puerta y fuera de puerta')
plt.xlabel('Tiros a puerta')
plt.ylabel('Tiros fuera de puerta')
plt.grid(True)
plt.show()
```

**Conclusiones: Disciplina:** El equipo con mayor cantidad de tarjetas amarillas fue **Italia**.

Capacidad Goleadora: Solo dos equipos lograron superar
la barrera de los 6 goles: **Alemania (10 goles)** y **España (12 goles).**

**Fair Play:** El equipo con menos faltas cometidas fue **Paises Bajos (Netherlands)** con **30 faltas concedidas.**

**Relacion de Tiros:** Al analizar el grafico de dispersion, no se observa una correlacion lineal clara entre **"Tiros  puerta" y "Tiros fuera de puerta"**.Esto indica que un mayor numero de tiros a puerta no implica por obligacion un aumento proporcional en tiros desviados o viceversa.
